```python
# function that can unpack the tuple and  use it to make judgements.
# function to judge the employee of the month on the basis of work hours employ has putin.
```

```python
def judge(tup):
    a = []
    b = []
    for x,y in tup:
        a.append(x)
        b.append(y)
    c = max(b)
    d = b.index(c)
    
    print(f"{a[d]} is the employee of the month with {c} hours of work")
    
```

```python
judge([('Abby',100),('Cassie', 800),('Billy',4000)])
```

```python
from random import shuffle

#shuffle list
def shuffle_list(mylist):
    shuffle(mylist)
    return mylist

#player guess
def player_guess():
    guess = ''
    while guess not in ['0','1','2']:
        guess = input("Pick a number: 0 ,1, or 2 : ")
    return int(guess)

#checking guess
def check_guess(mylist,guess):
    if mylist[guess] == 'o':
        print("Correct!")
        print(mylist)
    else:
        print("Wrong guess!")
        print(mylist)
    
mylist = ['','o','']
x = 1
while x :
    check_guess(shuffle_list(mylist),player_guess())
    x = int(input('wanna play more? (0/1) '))
```

```python
def grade(name,*scores):
    score = 0
    n = 0
    for x in scores : 
        score += x
        n += 1
    average_score = score/n
    print(f"{name} with average score of {average_score}")
```

```python
# global scope
def helper_bot():
    # Enclosing func
    # enclosing scope
    status = 'not helped the client yet'
    def use():
        # Nested func
        # local scope
        status = 'helped the client '
        print(status)
    use()
    print(status)
```

```python
helper_bot()
```

```python
# You can't retrieve variable value from nested func to enclosing func, but you can do opposite
# to retrive variable value from nested func to enclosing func, you can use nonlocal keyword

# global scope
def helper_bot():
    # Enclosing func
    # enclosing scope
    status = 'not helped the client yet'
    def use():
        # Nested func
        # local scope
        nonlocal status
        status = 'helped the client '
        print(status)
    use()
    print(status)
```

```python
helper_bot()
```

```python
def recursivefun_fact(n):
    if n<=1:
        return 1
    else:
        a = n*recursivefun_fact(n-1)
        return a
```

```python
x = recursivefun_fact(5) #calculating factorial
```

```python
print(x)
```

```python
def recursivefun_fibonacci(n):
    if n <= 2 :
        return n
    else:
        return recursivefun_fibonacci(n-1) + recursivefun_fibonacci(n-2)
```

```python
for i in range(10):
    print(i,recursivefun_fibonacci(i))
```

```python
def list_dir1(path):
    
    def dir_list(d):
        nonlocal tab_stop
        directories = os.listdir(d)
        for files in directories:
            current_dir = os.path.join(d,files)
            if os.path.isdir(current_dir):
                    print('\t'* tab_stop + 'directory' + files)
                    tab_stop += 1
                    dir_list(current_dir)
                    tab_stop -= 1
            else:
                print(files)
                
    tab_stop = 0
    
    if os.path.exists(path):
        print('Directory Listing of' + path)
        dir_list(path)
    else:
        print("Path doesn't exists")
```
